```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
caminho1 = r"C:\Projetos\LearningDataAnalysis\dados\mediaEnemPorMunicipio.csv"
```

```python
mediaEnemPorMunicipio = pd.read_csv(caminho1, sep=",", index_col=0)
```

```python
mediaEnemPorMunicipio.head()
```

```python
caminho2 = r"C:\Projetos\LearningDataAnalysis\dados\atlasMunicipios.csv"
```

```python
atlasHumano = pd.read_csv(caminho2, sep=",", index_col=0)
```

```python
atlasHumano.head()
```

```python
enem_idhm = pd.merge(mediaEnemPorMunicipio, atlasHumano, how='left', left_on=['CO_MUNICIPIO_ESC'], right_on=['Codmun7'])
```

```python
enem_idhm.head()
```

```python
mediaEnemIDHM = enem_idhm.filter(['CO_MUNICIPIO_ESC','NO_MUNICIPIO_ESC', 'SG_UF_ESC	',
                                  'CO_UF_ESC', 'NU_NOTA_MT', 'NU_NOTA_LC',
                                  'NU_NOTA_CH', 'NU_NOTA_CN', 'NU_NOTA_REDACAO',
                                  'IDHM', 'IDHM_E', 'IDHM_L', 'IDHM_R'])
mediaEnemIDHM 
```

```python
mediaEnemIDHM.plot.scatter('NU_NOTA_REDACAO', 'IDHM')
```
