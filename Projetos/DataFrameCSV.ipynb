```python
import pandas as pd
import matplotlib.pyplot as plt
import re
```

```python
caminho = r"C:\Projetos\LearningDataAnalysis\dados\mediaEnemPorMunicipio.csv"
```

```python
microdadosEnem = pd.read_csv(caminho, sep=",", encoding='ISO-8859-1')
```

```python
microdadosEnem.columns.values
```

```python
mediaEnemPorMunicipio = (
    microdadosEnem
    .groupby(['CO_MUNICIPIO_ESC', 'NO_MUNICIPIO_ESC', 'SG_UF_ESC', 'CO_UF_ESC'], as_index=False)
    .mean(numeric_only=True)
    .sort_values(by='SG_UF_ESC')
)
```

```python
mediaEnemPorMunicipio
```

```python
mediaEnemPorMunicipio.to_csv("mediaEnemPorMunicipio.csv", index=False)
```
