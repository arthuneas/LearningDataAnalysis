```python
import pandas as pd
import matplotlib.pyplot as plt
import re
```

```python
caminho = r"C:\Projetos\LearningDataAnalysis\dados\mediaEnemPorMunicipio.csv"
```

```python
microdadosEnem = pd.read_csv(caminho, sep=";", encoding='ISO-8859-1')
```

```python
microdadosEnem.columns.values
```

```python
microdadosEnemSelecionado = microdadosEnem.filter(items=['CO_MUNICIPIO_ESC', 'NO_MUNICIPIO_ESC', 'SG_UF_ESC', 'CO_UF_ESC', 'NU_NOTA_MT','NU_NOTA_LC', 'NU_NOTA_CH', 'NU_NOTA_CN','NU_NOTA_REDACAO'])
```

```python
mediaEnemPorMunicipio = microdadosEnemSelecionado.groupby(['CO_MUNICIPIO_ESC', 'NO_MUNICIPIO_ESC', 'SG_UF_ESC',  'CO_UF_ESC'], as_index=False).mean(numeric_only=True)#.sort_values(by = 'SG_UF_ESC')
```

```python
mediaEnemPorMunicipio.to_csv("mediaEnemPorMunicipio.csv")
```
