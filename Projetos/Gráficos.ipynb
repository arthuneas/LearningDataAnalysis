```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.plot([1,2 ,3 ,4]) #pontos
plt.show()
```

```python
plt.plot([1,2 ,3 ,4], "ro") #pontos
plt.show()
```

```python
x = [1, 2 ,3 ,4]
y = np.arange(0, 10, 0.1) #os dois priemiros parametros são limites e o ultimo: incrementador
fig1, f1_axes = plt.subplots(ncols=2, nrows=2, figsize=(15, 10))
fig1.suptitle("Intro aos Gráficos Matplotlib", size=20)

f1_axes[0,0].plot(x)
f1_axes[0,0].set_title("gráfico de reta")
f1_axes[0,0].set_xlabel("label x")
f1_axes[0,0].set_ylabel("label y")

f1_axes[0,1].plot(np.tan(y))
f1_axes[0,1].set_title("gráfico de tangente")

f1_axes[1,0].plot(np.sin(y), label="sen(x)", color="pink")
f1_axes[1,0].plot(np.cos(y), label="cos(x)", color="olive")
f1_axes[1,0].set_title("gráfico de sen + cos")
f1_axes[1,0].legend()

f1_axes[1,1].plot(np.cos(y))
f1_axes[1,1].set_title("gráfico de cosseno")
```
