```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
caminho_municipios = r"C:\Projetos\LearningDataAnalysis\dados\atlasMunicipios.csv"
```

```python
caminho_dados = r"C:\Projetos\LearningDataAnalysis\dados\mediaEnemPorMunicipio.csv"
```

```python
municipios = pd.read_csv(caminho_municipios, sep=",", encoding='utf-8')
```

```python
municipios = municipios[municipios['ANO'] == 2010] #ESCOLHENDO O ANO DE ANALISE
```

```python
dadosEnem = pd.read_csv(caminho_dados, sep=",", index_col=0, encoding='utf-8')
```

```python
enem_municipios = pd.merge(dadosEnem, municipios, how='left', left_on=['CO_MUNICIPIO_ESC'], right_on=['Codmun7'])
```

```python
mediaEnemIDHM = enem_municipios.filter(['CO_MUNICIPIO_ESC','NO_MUNICIPIO_ESC', 'SG_UF_ESC',
                                      'CO_UF_ESC', 'NU_NOTA_MT', 'NU_NOTA_LC',
                                      'NU_NOTA_CH', 'NU_NOTA_CN', 'NU_NOTA_REDACAO',
                                      'IDHM', 'IDHM_E', 'IDHM_L', 
                                    'IDHM_R', 'LONG', 'LAT', 
                                    'UF', 'Codmun7', 'Municipio'])
mediaEnemIDHM 
```

```python
mediaEnemIDHM.plot.scatter(y='LAT', x='LONG',
                           c=mediaEnemIDHM['NU_NOTA_REDACAO'],
                           cmap='RdYlGn',
                           figsize=(10, 10),
                          title = 'Nota da Redação')
```

```python
fig, ax = plt.subplots(2, 2, figsize=(20, 16), sharex='col', sharey='row')
fig.suptitle('Desempenho de Nota nas Áreas de Conhecimento do ENEM pelo Território Brasileiro', fontsize=20, fontweight='bold')

mediaEnemIDHM.plot.scatter(y='LAT', x='LONG',
                           c=mediaEnemIDHM['NU_NOTA_LC'],
                           cmap='RdYlGn',
                          title = 'Nota em Linguagens', ax=ax[0,0])



mediaEnemIDHM.plot.scatter(y='LAT', x='LONG',
                           c=mediaEnemIDHM['NU_NOTA_CH'],
                           cmap='RdYlGn',
                          title = 'Nota em Ciências Humanas', ax=ax[0,1])


mediaEnemIDHM.plot.scatter(y='LAT', x='LONG',
                           c=mediaEnemIDHM['NU_NOTA_CN'],
                           cmap='RdYlGn',
                          title = 'Nota em Ciências da Natureza', ax=ax[1,0])


mediaEnemIDHM.plot.scatter(y='LAT', x='LONG',
                           c=mediaEnemIDHM['NU_NOTA_MT'],
                           cmap='RdYlGn',
                          title = 'Nota em Matemática', ax=ax[1,1])
```

```python
fig, ax = plt.subplots(2, 2, figsize=(20, 16), sharex='col', sharey='row')
fig.suptitle('Análise do IDHM e seus Conjuntos no Terrtório Brasileiro', fontsize=20, fontweight='bold')

mediaEnemIDHM.plot.scatter(y='LAT', x='LONG',
                           c=mediaEnemIDHM['IDHM'],
                           cmap='coolwarm',
                          title = 'IDHM', ax=ax[0,0])



mediaEnemIDHM.plot.scatter(y='LAT', x='LONG',
                           c=mediaEnemIDHM['IDHM_E'],
                           cmap='coolwarm',
                          title = 'IDHM de Escolaridade', ax=ax[0,1])


mediaEnemIDHM.plot.scatter(y='LAT', x='LONG',
                           c=mediaEnemIDHM['IDHM_L'],
                           cmap='coolwarm',
                          title = 'IDHM de Longevidade', ax=ax[1,0])


mediaEnemIDHM.plot.scatter(y='LAT', x='LONG',
                           c=mediaEnemIDHM['IDHM_R'],
                           cmap='coolwarm',
                          title = 'IDHM de Renda', ax=ax[1,1])
```
